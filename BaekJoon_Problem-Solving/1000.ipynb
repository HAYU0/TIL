# 백준풀이

#### No.1000

[문제 링크](https://www.acmicpc.net/problem/1000)

### 🐣 내 풀이

```python
A, B = input().split()
print(int(A)+int(B))
```

### 💡 다른 풀이

```python
A, B = input().split()	# 입력되는 문자를 input()함수로 입력받고 split()함수로 나누어 A,B 변수에 저장

print(int(A)+int(B))	# int() 함수로 A와 B를 정수로 변환 하고 두수의 합을 출력
```
